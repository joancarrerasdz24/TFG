```python
from scipy.signal import butter, detrend, filtfilt
import numpy as np
```

Crear una señal de ejemplo

Crear una señal con una tendencia lineal

```python
t = np.linspace(0, 1, 100)
signal = 2 * t + np.random.normal(0, 0.1, 100)
```

Detrender la señal

```python
detrended_signal = detrend(signal)
```

Visualizar la señal original y la señal detrended

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label='Señal Original')
plt.plot(t, detrended_signal, label='Señal Detrended')
plt.legend()
plt.show()
```

```python
signal = detrended_signal
```

Diseñar un filtro Butterworth

```python
order = 4  # Orden del filtro
cutoff_freq = 2  # Frecuencia de corte
b, a = butter(order, cutoff_freq, fs=10, btype='low')
```

Aplicar el filtro a la señal

```python
filtered_signal = filtfilt(b, a, signal)
```

Visualizar la señal original y la señal filtrada

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label='Señal Original')
plt.plot(t, filtered_signal, label='Señal Filtrada')
plt.legend()
plt.show()
```
